```python
import numpy as np
from scipy.cluster import hierarchy
import pandas as pd
data=pd.read_csv("/content/Iris (1).csv",index_col=0)
```

```python
data
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
x=data[["SepalLengthCm",	"SepalWidthCm",	"PetalLengthCm",	"PetalWidthCm"]]
```

```python
pca=pca.fit_transform(x)
```

```python
df=pd.DataFrame(pca,columns=["PC1","PC2"])
```

```python
df
```

```python
y=hierarchy.linkage(df,"single")
z=hierarchy.dendrogram(y)
```

```python
y=hierarchy.linkage(df,"complete")
z=hierarchy.dendrogram(y)
```

```python
y=hierarchy.linkage(df,"average")
z=hierarchy.dendrogram(y)
```
