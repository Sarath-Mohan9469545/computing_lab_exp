```python
import pandas as pd
data=pd.read_csv("/content/Iris.csv")
data
```

```python
x=data[["SepalLengthCm",	"SepalWidthCm",	"PetalLengthCm",	"PetalWidthCm"]]
x
```

```python
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
pca=pca.fit_transform(x)
```

```python
new_df=pd.DataFrame(pca,columns=["PC1","PC2"])
```

```python
new_df
```

```python
ML=DBSCAN()
ML=ML.fit(new_df)
labels=ML.labels_
```

```python
labels
```

```python
import matplotlib.pyplot as plt
xaxis=new_df["PC1"]
yaxis=new_df["PC2"]
plt.scatter(xaxis,yaxis,c=labels,cmap="rainbow")
```

```python
data2=pd.read_csv("/content/heart_failure_clinical_records_dataset.csv")
```

```python
data2
```

```python
x=data2[["age",	"anaemia",	"creatinine_phosphokinase",	"diabetes",	"ejection_fraction",	"high_blood_pressure",	"platelets"	,"serum_creatinine"	,"serum_sodium",	"sex",	"smoking"	,"time"]]
```

```python
x
```

```python
pca=PCA(n_components=2)
pca=pca.fit_transform(x)
new_df=pd.DataFrame(pca,columns=["PC1","PC2"])
new_df
```

```python
ML=DBSCAN(eps=20, min_samples=2)
ML=ML.fit(new_df)
labels=ML.labels_
```

```python
labels
```

```python
import matplotlib.pyplot as plt
xaxis=new_df["PC1"]
yaxis=new_df["PC2"]
plt.scatter(xaxis,yaxis,c=labels,cmap="rainbow")
```
