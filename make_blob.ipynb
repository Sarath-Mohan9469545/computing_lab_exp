```python
from sklearn.datasets import make_blobs
```

```python
x, y = make_blobs(n_samples=2000, centers=2, n_features=2)
```

```python
x
```

```python
y
```

```python
import pandas as pd
new_df=pd.DataFrame(x,columns=["X_axis","Y_axis"])
```

```python
new_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
X_train
```

```python
X_train=pd.DataFrame(X_train,columns=["X_axis","Y_axis"])
```

```python
X_train
```

```python
import matplotlib.pyplot as plt
x_axis=X_train["X_axis"]
y_axis=X_train["Y_axis"]
plt.scatter(x_axis,y_axis,s=10, cmap="y")
plt.xlabel("X")
plt.ylabel("Y")
plt.rcParams["figure.figsize"]=(10,8)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(X_train)
```

```python
c=kmeans.cluster_centers_
c
```

```python
import matplotlib.pyplot as plt
x_axis=X_train["X_axis"]
y_axis=X_train["Y_axis"]
labels=kmeans.labels_
plt.scatter(x_axis,y_axis, s=20,c=labels,cmap="rainbow")
plt.scatter(c[0][0],c[0][1],s=100,color="orange")
plt.scatter(c[1][0],c[1][1],s=100,color="green")

plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
X_test=pd.DataFrame(X_test,columns=["X_axis","Y_axis"])
```

```python
res1=kmeans.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(res1, y_test)
```
